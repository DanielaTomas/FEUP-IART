## Data pre-processing

### NA values

```python
print(data.isna().sum())
print()
print("There are " + str((data.isna().sum().sum())) + " NA values")
```

### Duplicate values

```python
length = len(data)
data.drop_duplicates(inplace=True)

print("Original dataset length: " + str(length))
print("Number of duplicates: " + str(length - len(data)))
```
